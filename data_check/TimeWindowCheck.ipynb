```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib as mpl
```

```python
from matplotlib import ticker
import matplotlib as mpl
label_size = 15
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size
```

```python
keys = ["nutau", "nue", "numu", "2012", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021"]
weights_files = [334, 602, 1518]
times = {}
weights = {}
for i,key in enumerate(keys):
    input_file = "/mnt/home/micall12/LowEnergyNeuralNetwork/time_window_%s.hdf5"%key
    f = h5py.File(input_file, 'r')
    times[key] = f['time_window'][:]
    if key == "nue" or key =="numu" or key == "nutau":
        weights[key] = f['weights'][:]/weights_files[i]
    f.close()
    del f
```

```python
times["MC"] = np.concatenate((times[keys[0]], times[keys[1]], times[keys[2]]))
weights["MC"] = np.concatenate((weights[keys[0]], weights[keys[1]], weights[keys[2]]))
times["data"] = np.concatenate((times[keys[3]], times[keys[4]], times[keys[5]],
                                times[keys[6]], times[keys[7]], times[keys[8]],
                                times[keys[9]], times[keys[10]], times[keys[11]],
                                times[keys[12]], times[keys[13]]))
```

```python
for key in keys:
    print(key,len(times[key]))
print("MC",len(times["MC"]))
print("data",len(times["data"]))
```

```python
print(max(times["data"]),max(times["MC"]))
```

```python
amin=10000
amax=18000
bins=20
bin_size = (amax-amin)/20
data_hist, data_bins, p = plt.hist(times["data"],range=[amin,amax],bins=bins)
plt.yscale('log')
mid_bins = data_bins[:-1] + bin_size/2
```

```python
list_times = []
list_weights = []
for i in range(0,3):
    list_times.append(times[keys[i]])
    list_weights.append(weights[keys[i]]*1000)
plt.hist(list_times,weights=list_weights,stacked=True,label=keys[:3],range=[amin,amax],bins=bins)
plt.legend(fontsize=15)
plt.xlabel("Time Range (ns)",fontsize=20)
plt.yscale('log')
plt.ylabel("Rate (mHz)",fontsize=20)
```

```python
plt.hist(list_times,weights=list_weights,stacked=True,label=keys[:3],range=[amin,amax],bins=bins)
plt.plot(mid_bins,data_hist,"k.",markersize=10,label="data")
plt.legend(fontsize=15)
plt.xlabel("Time Range (ns)",fontsize=20)
plt.yscale('log')
#plt.ylabel("Rate (mHz)",fontsize=20)
```
